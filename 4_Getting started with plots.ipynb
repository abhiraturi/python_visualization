Generate a sine wave

```python
x=np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x))

plt.xlabel('x',fontsize=15,color='green')
plt.ylabel('Sin(x)',fontsize=15,color='green')
```

```python
plt.tick_params(axis='y',
               color='red',
               labelcolor='blue',
               labelsize='xx-large')
```

```python
plt.tick_params(axis='x',
               bottom=False,
               labelbottom=False)
```

```python
plt.plot([2,4,6,8],
        [4,8,12,16])

plt.ylim(ymin=0)
plt.xlim(xmin=0)
```

Add legend

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend()
plt.title('Sine Curve')
```

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend()
plt.title('Sine Curve')
plt.xlim(1,5)
```

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend()
plt.title('Sine Curve')
plt.xlim(1,5)
plt.ylim(-1 ,0.5)
```

Invert the axes

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend(loc='upper center')
plt.title('Sine Curve')

plt.xlim(5,1)
plt.ylim(-1,0.5)
plt.savefig("C:/Users/HP/Downloads/Sinecurve.jpg")
```

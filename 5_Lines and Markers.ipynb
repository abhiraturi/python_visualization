```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x))
plt.xlabel('x',fontsize=15,color='green')
plt.xlabel('Sin(x)',fontsize=15,color='green')
```

```python
plt.plot(x,np.sin(x),label='Sine curve',color='green')
plt.xlabel('x',fontsize=15,color='green')
plt.ylabel('Sin(x)',fontsize=15,color='green')

plt.plot(x,np.cos(x),label='Cosine Curve',color='m')
plt.legend(loc='upper right')
plt.title('Sine and Cosine Curves')
```

```python
xarray=[2,4,6,8]
yarray=[4,8,12,16]

plt.plot(xarray,yarray,marker='D')

# adding a text
plt.text(2,4,'this text starts at point(2,4)')

plt.text(8,16,'this text ends at point(8,16)',horizontalalignment='right')
```

```python
plt.scatter(xarray,yarray)

for x,y in zip(xarray,yarray):
    
    label= f"({x},{y})"
    print(label)
    
    plt.annotate(label,
                (x,y),
                textcoords="offset points",
                xytext=(0,10),
                ha='right')
```

```python
plt.plot(xarray,yarray,'bo-')
plt.grid(True)

for x,y in zip(xarray,yarray):
    
    label= "{:.2f}" .format(y)
   
    
    plt.annotate(label,
                (x,y),
                textcoords="offset points",
                xytext=(0,10),
                ha='center')
 
```
